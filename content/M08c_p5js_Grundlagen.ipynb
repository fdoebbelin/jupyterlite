## ‚úÖ L√∂sung 1: Dein erstes Kunstwerk

**Anforderung:** Ein Kunstwerk aus mindestens 3 verschiedenen Formen mit mindestens 3 Farben.

### Beispiel-L√∂sung 1: Abstrakte Kunst

````p5js
function setup() {
  createCanvas(400, 300);
}

function draw() {
  // Hintergrund
  background(240);
  
  // Rotes Rechteck (unten links)
  fill(255, 0, 0);        // Rot
  stroke(0);              // Schwarzer Umriss
  strokeWeight(2);
  rect(30, 200, 120, 80);
  
  // Blauer Kreis (oben rechts)
  fill(0, 0, 255);        // Blau
  circle(320, 80, 100);
  
  // Gr√ºnes Dreieck (Mitte)
  fill(0, 200, 0);        // Gr√ºn
  noStroke();             // Kein Umriss
  triangle(200, 50, 100, 250, 300, 250);
  
  // Gelbe Linie (quer durch)
  stroke(255, 255, 0);    // Gelb
  strokeWeight(5);
  line(50, 150, 350, 150);
}
````

## ‚úÖ L√∂sung 2: Das Haus zeichnen

**Anforderung:** Ein Haus mit Dach, W√§nden, Fenstern und T√ºr.

### Vollst√§ndige L√∂sung

````p5js
function setup() {
  createCanvas(500, 400);
}

function draw() {
  background(135, 206, 235);  // Hellblauer Himmel
  
  // ===== W√ÑNDE (Rechteck) =====
  fill(210, 180, 140);         // Beige
  stroke(0);                   // Schwarzer Umriss
  strokeWeight(2);
  rect(150, 150, 200, 150);    // x=150, y=150, Breite=200, H√∂he=150
  
  // ===== DACH (Dreieck) =====
  fill(139, 69, 19);           // Braun
  stroke(0);
  strokeWeight(2);
  // Dreieck: (x1, y1), (x2, y2), (x3, y3)
  // Dach oben auf den W√§nden positionieren
  triangle(150, 150,    // Linke untere Ecke des Daches
           250, 50,     // Spitze des Daches (Mitte oben)
           350, 150);   // Rechte untere Ecke des Daches
  
  // ===== LINKES FENSTER (oben links) =====
  fill(135, 206, 235);  // Hellblaues Glas (wie Himmel)
  stroke(0);
  strokeWeight(2);
  rect(180, 170, 40, 40);  // x=180, y=170, Breite=40, H√∂he=40
  
  // Fensterkreuz (schwarze Linien)
  stroke(0);
  line(200, 170, 200, 210);  // Vertikale Linie
  line(180, 190, 220, 190);  // Horizontale Linie
  
  // ===== RECHTES FENSTER (oben rechts) =====
  fill(135, 206, 235);  // Hellblaues Glas
  stroke(0);
  strokeWeight(2);
  rect(280, 170, 40, 40);  // x=280, y=170, Breite=40, H√∂he=40
  
  // Fensterkreuz
  stroke(0);
  line(300, 170, 300, 210);  // Vertikal
  line(280, 190, 320, 190);  // Horizontal
  
  // ===== T√úR (Mitte unten) =====
  fill(101, 67, 33);  // Dunkelbraun
  stroke(0);
  strokeWeight(2);
  rect(220, 240, 60, 60);  // x=220, y=240, Breite=60, H√∂he=60
  
  // T√ºrklinke (kleiner Kreis)
  fill(255, 215, 0);  // Gold
  circle(275, 270, 8);  // x=275, y=270, Durchmesser=8
  
  // ===== SCHORNSTEIN (optional) =====
  fill(200, 100, 100);  // R√∂tlich
  stroke(0);
  strokeWeight(2);
  rect(280, 70, 30, 70);  // x=280, y=70, Breite=30, H√∂he=70
  
  // ===== GRAS (Rechteck unten) =====
  fill(34, 139, 34);  // Dunkelgr√ºn
  noStroke();
  rect(0, 300, 500, 100);
}
````

**Erkl√§rung der Positionen:**

```
        Dach (Dreieck)
         /   \
    [Fenster][Fenster]
        
       [T√úR]
    
    ___Gras___
```

**Wichtige Erkenntnisse:**
- Das Dach-Dreieck wird mit `triangle()` auf den W√§nden positioniert

## ‚úÖ L√∂sung 3: Koordinatensystem erkunden

**Anforderung:** Kreuze an 5 Positionen (4 Ecken + Mitte).

### Vollst√§ndige L√∂sung

````p5js
function setup() {
  createCanvas(400, 400);
}

function draw() {
  background(220);
  
  // Hilfsfunktion: Zeichne ein Kreuz an einer Position
  function drawCross(x, y, color) {
    stroke(color);
    strokeWeight(2);
    // Horizontale Linie
    line(x - 15, y, x + 15, y);
    // Vertikale Linie
    line(x, y - 15, x, y + 15);
  }
  
  // Oben links (0, 0)
  drawCross(0, 0, 255, 0, 0);        // Rot
  
  // Oben rechts (400, 0)
  drawCross(400, 0, 0, 255, 0);      // Gr√ºn
  
  // Unten links (0, 400)
  drawCross(0, 400, 0, 0, 255);      // Blau
  
  // Unten rechts (400, 400)
  drawCross(400, 400, 255, 255, 0);  // Gelb
  
  // Mitte (200, 200)
  drawCross(200, 200, 255, 0, 255);  // Magenta
  
  // Zus√§tzliche Hilfslinien (optional)
  stroke(200);
  strokeWeight(1);
  line(200, 0, 200, 400);    // Vertikale Mittellinie
  line(0, 200, 400, 200);    // Horizontale Mittellinie
  
  // Text-Beschriftung (optional, ben√∂tigt text() Funktion)
  fill(0);
  textSize(12);
  text("(0,0)", 5, 15);
  text("(400,0)", 350, 15);
  text("(0,400)", 5, 395);
  text("(400,400)", 330, 395);
  text("(200,200)", 210, 195);
}
````

## ‚úÖ L√∂sung 4: Muster mit Schleifen

**Anforderung:** 10 Kreise in einer Reihe mit `for`-Schleife.

### Vollst√§ndige L√∂sung

````p5js
function setup() {
  createCanvas(400, 400);
}

function draw() {
  background(220);
  
  // 10 Kreise in horizontaler Reihe
  for (let i = 0; i < 10; i++) {
    // x-Position berechnen
    let x = 20 + i * 40;  // Start bei 20, dann alle 40 Pixel ein Kreis
    let y = 200;          // Alle Kreise auf gleicher H√∂he
    
    // Farbe basierend auf Index
    let r = i * 25;       // Rot erh√∂ht sich
    let g = 100;          // Gr√ºn bleibt gleich
    let b = 200;          // Blau bleibt gleich
    
    fill(r, g, b);
    stroke(0);
    strokeWeight(1);
    circle(x, y, 30);
  }
}
````

````p5js
%show
````

**Ergebnis:** 10 Kreise mit Farbverlauf von dunklem Blau zu heller Mischung.

### Variante 2: Mit Regenbogenfarben

````p5js
function setup() {
  createCanvas(400, 400);
  colorMode(HSB);  // HSB = Hue, Saturation, Brightness (Farbton, S√§ttigung, Helligkeit)
}

function draw() {
  background(220);
  
  for (let i = 0; i < 10; i++) {
    let x = 20 + i * 40;
    let y = 200;
    
    // Farbton wechselt wie ein Regenbogen
    let hue = (i / 10) * 360;  // 0 - 360 Grad
    
    fill(hue, 255, 255);
    noStroke();
    circle(x, y, 30);
  }
}
````

### Variante 3: Vertikale Reihe

````p5js
function setup() {
  createCanvas(400, 400);
}

function draw() {
  background(220);
  
  // 10 Kreise vertikal
  for (let i = 0; i < 10; i++) {
    let x = 200;
    let y = 20 + i * 40;
    
    let brightness = i * 25;  // Von dunkel zu hell
    fill(brightness);
    noStroke();
    circle(x, y, 30);
  }
}
````

## ‚úÖ L√∂sung 5: Farbverlauf-Gitter

**Anforderung:** 5√ó5 Gitter mit Farbverlauf von oben links (dunkelblau) zu rechts unten (hellrot).

### Vollst√§ndige L√∂sung

````p5js
function setup() {
  createCanvas(300, 300);
}

function draw() {
  background(220);
  noStroke();
  
  // Doppelschleife: Zeilen und Spalten
  for (let row = 0; row < 5; row++) {
    for (let col = 0; col < 5; col++) {
      // Position berechnen
      let x = col * 60;      // x-Position (Spalte √ó Gr√∂√üe)
      let y = row * 60;      // y-Position (Reihe √ó Gr√∂√üe)
      
      // Farbe basierend auf Position
      // Rot: nimmt nach rechts zu (von 0 zu 255)
      let r = (col / 5) * 255;
      
      // Gr√ºn: middle (konstant)
      let g = 100;
      
      // Blau: nimmt nach unten ab (von 255 zu 0)
      let b = 255 - (row / 5) * 255;
      
      // Rechteck zeichnen
      fill(r, g, b);
      rect(x, y, 60, 60);
      
      // Optional: Umrisse f√ºr bessere Sichtbarkeit
      stroke(0);
      strokeWeight(0.5);
    }
  }
}
````

````p5js
%show
````

**Erkl√§rung der Farbberechnung:**

| Position | Row | Col | R-Wert | G-Wert | B-Wert | Farbe |
|----------|-----|-----|--------|--------|--------|-------|
| Oben Links | 0 | 0 | 0 | 100 | 255 | Dunkelblau |
| Oben Rechts | 0 | 4 | 255 | 100 | 255 | Hellblau |
| Unten Links | 4 | 0 | 0 | 100 | 0 | Gr√ºn |
| Unten Rechts | 4 | 4 | 255 | 100 | 0 | Rot/Gelb |

### Alternative: Mit Division-Operator

````p5js
function setup() {
  createCanvas(300, 300);
}

function draw() {
  background(220);
  
  for (let row = 0; row < 5; row++) {
    for (let col = 0; col < 5; col++) {
      let x = col * 60;
      let y = row * 60;
      
      let r = col * 51;       // 255 / 5 = 51
      let g = 100;
      let b = 255 - row * 51; // 255 - (row * 51)
      
      fill(r, g, b);
      rect(x, y, 60, 60);
    }
  }
}
````

````p5js
%show
````

**Wichtige Erkenntnisse:**
- Verschachtelte Schleifen f√ºr 2D-Gitter
- Position: `x = col * 60`, `y = row * 60`
- Farbe √§ndert sich mit Row und Col
- Division durch Anzahl f√ºr sanfte √úberg√§nge

## ‚úÖ L√∂sung 7: Bewegter Ball

**Anforderung:** Ball bewegt sich diagonal und prallt an Kanten ab.

### Vollst√§ndige L√∂sung

````p5js
// Globale Variablen (au√üerhalb von setup() und draw())
let x = 200;
let y = 150;
let vx = 3;    // Geschwindigkeit x-Richtung
let vy = 2;    // Geschwindigkeit y-Richtung
let ballColor;
let ballRadius = 15;  // Radius des Balls (Durchmesser / 2)

function setup() {
  createCanvas(400, 300);
  // Zuf√§llige Farbe beim Start
  ballColor = color(random(255), random(255), random(255));
}

function draw() {
  background(220);
  
  // ===== BEWEGUNG =====
  x = x + vx;
  y = y + vy;
  
  // ===== RANDERKENNUNG UND ABPRALLEN =====
  // Linke und rechte Kante
  if (x - ballRadius < 0 || x + ballRadius > 400) {
    vx = -vx;  // Richtung umkehren
  }
  
  // Obere und untere Kante
  if (y - ballRadius < 0 || y + ballRadius > 300) {
    vy = -vy;  // Richtung umkehren
  }
  
  // ===== ZEICHNEN =====
  fill(ballColor);
  stroke(0);
  strokeWeight(2);
  circle(x, y, ballRadius * 2);  // Durchmesser = 2 √ó Radius
}
````

````p5js
%show
````

**Erkl√§rung:**

1. **Variablen definieren**: `x`, `y` (Position), `vx`, `vy` (Geschwindigkeit)
2. **Bewegung**: `x = x + vx` und `y = y + vy` √§ndern die Position
3. **Randerkennung**: √úberpr√ºfe, ob Ball die Kante √ºberschreitet
4. **Abprallen**: `vx = -vx` dreht die Geschwindigkeit um

### Alternative: Mit zus√§tzlicher Geschwindigkeit

````p5js
let x = 200;
let y = 150;
let vx = 4;
let vy = 3;
let ballColor;

function setup() {
  createCanvas(400, 300);
  ballColor = color(255, 100, 100);  // Rot
}

function draw() {
  background(220);
  
  // Bewegung
  x += vx;
  y += vy;
  
  // Abprallen (vereinfacht)
  if (x < 15 || x > 385) vx *= -1;
  if (y < 15 || y > 285) vy *= -1;
  
  // Zeichnen
  fill(ballColor);
  circle(x, y, 30);
}
````

````p5js
%show
````

**Wichtige Erkenntnisse:**
- Globale Variablen speichern Zustand zwischen Frames
- `x += vx` ist das gleiche wie `x = x + vx`
- Randerkennung: `x - radius < 0` oder `x + radius > width`
- `vx *= -1` ist das gleiche wie `vx = -vx`

## ‚úÖ L√∂sung 8: Zuf√§lliges Stippen

**Anforderung:** 50 zuf√§llige Punkte mit zuf√§lligen Farben, jede Sekunde neu.

### Vollst√§ndige L√∂sung

````p5js
function setup() {
  createCanvas(400, 400);
}

function draw() {
  background(220);
  noStroke();
  
  // 50 zuf√§llige Punkte zeichnen
  for (let i = 0; i < 50; i++) {
    // Zuf√§llige Position
    let x = random(400);        // Zahl zwischen 0 und 400
    let y = random(400);        // Zahl zwischen 0 und 400
    
    // Zuf√§lliger Durchmesser (zwischen 5 und 30)
    let diameter = random(5, 30);
    
    // Zuf√§llige Farbe
    let r = random(255);
    let g = random(255);
    let b = random(255);
    
    // Zeichnen
    fill(r, g, b);
    circle(x, y, diameter);
  }
}
````

````p5js
%show
````

**Ergebnis:** Jedes Frame werden 50 neue zuf√§llige Punkte gezeichnet ‚Üí animiertes Kunstwerk!

### Alternative: Mit Verz√∂gerung (nur alle 60 Frames erneuern)

````p5js
function setup() {
  createCanvas(400, 400);
}

function draw() {
  // Nur alle 60 Frames (‚âà 1 Sekunde bei 60 FPS) erneuern
  if (frameCount % 60 === 0) {
    background(220);
    
    for (let i = 0; i < 50; i++) {
      let x = random(400);
      let y = random(400);
      let diameter = random(5, 30);
      
      fill(random(255), random(255), random(255));
      circle(x, y, diameter);
    }
  }
}
````

````p5js
%show
````

### Alternative: Mit Speicherung der Punkte

````p5js
let points = [];

function setup() {
  createCanvas(400, 400);
  generatePoints();  // Erste Punkte generieren
}

function draw() {
  background(220);
  noStroke();
  
  // Alle Punkte zeichnen
  for (let point of points) {
    fill(point.r, point.g, point.b);
    circle(point.x, point.y, point.diameter);
  }
  
  // Neue Punkte alle 60 Frames
  if (frameCount % 60 === 0) {
    generatePoints();
  }
}

function generatePoints() {
  points = [];  // Alte Punkte l√∂schen
  
  for (let i = 0; i < 50; i++) {
    points.push({
      x: random(400),
      y: random(400),
      diameter: random(5, 30),
      r: random(255),
      g: random(255),
      b: random(255)
    });
  }
}
````

````p5js
%show
````

**Wichtige Erkenntnisse:**
- `random(400)` gibt eine Zahl zwischen 0 und 400
- `random(5, 30)` gibt eine Zahl zwischen 5 und 30
- `frameCount % 60` ist 0 alle 60 Frames
- Objekte k√∂nnen mehrere Eigenschaften speichern

## üß† H√§ufige Fehler und L√∂sungen

### Fehler 1: Formen sind nicht sichtbar

**Problem:** `fill()` wurde mit der gleichen Farbe wie `background()` gesetzt.

**L√∂sung:**

````p5js
background(255);      // Wei√ü
fill(200);             // Grau (nicht wei√ü!)
rect(50, 50, 100, 100);
````

### Fehler 2: Reihenfolge der Funktionen falsch

**Problem:** `background()` wird nach den Formen aufgerufen.

````p5js
function draw() {
  circle(200, 200, 50);
  background(220);  // FALSCH! L√∂scht den Kreis
}
````

**L√∂sung:**

````p5js
function draw() {
  background(220);  // RICHTIG! Zuerst Hintergrund
  circle(200, 200, 50);
}
````

### Fehler 3: Variablen nicht globalisiert

**Problem:** Variablen werden in `draw()` definiert und verlieren ihren Wert.

````p5js
function draw() {
  let x = 0;       // Wird jedes Frame auf 0 zur√ºckgesetzt!
  x = x + 5;
  circle(x, 200, 30);
}
````

**L√∂sung:**

````p5js
let x = 0;         // GLOBAL! Au√üerhalb von draw()

function draw() {
  x = x + 5;       // Jetzt wird x gr√∂√üer
  circle(x, 200, 30);
}
````

### Fehler 4: Koordinaten au√üerhalb des Canvas

**Problem:** Der Punkt wird nicht gezeichnet, weil die Koordinaten au√üerhalb sind.

````p5js
createCanvas(300, 300);
circle(400, 400, 50);  // Au√üerhalb! (300√ó300 Canvas)
````

**L√∂sung:** √úberpr√ºfe die Koordinaten:
- x sollte zwischen 0 und `width` sein
- y sollte zwischen 0 und `height` sein

### Fehler 5: `setup()` wird nicht aufgerufen

**Problem:** `createCanvas()` wird nicht aufgerufen.

````p5js
function draw() {
  createCanvas(400, 300);  // FALSCH! Wird jedes Frame aufgerufen
}
````

**L√∂sung:**

````p5js
function setup() {
  createCanvas(400, 300);  // RICHTIG! Wird nur einmal aufgerufen
}
````

## üéì Zusammenfassung der Konzepte

| Konzept | Erkl√§rung | Beispiel |
|---------|-----------|---------|
| **setup()** | Wird einmal am Anfang aufgerufen | `createCanvas()`, Startfarbe |
| **draw()** | Wird jedes Frame aufgerufen | Bewegung, Animation, Interaktion |
| **Koordinaten** | (0,0) oben links, x nach rechts, y nach unten | `circle(200, 150, 50)` |
| **RGB-Farben** | Rot, Gr√ºn, Blau (0-255 jeweils) | `fill(255, 0, 0)` = Rot |
| **Formen** | rect, circle, triangle, line | `circle(x, y, diameter)` |
