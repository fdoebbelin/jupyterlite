## ðŸ““ Praktische Demonstrationen in Jupyter

### Demonstration 1: Ein einfaches Gesicht

````p5js
function setup() {
  createCanvas(300, 400);
  background(255);
}

function draw() {
  // Kopf (gelb)
  fill(255, 255, 0);
  stroke(0);
  strokeWeight(3);
  circle(150, 150, 150);
  
  // Linkes Auge
  fill(0);
  circle(120, 120, 20);
  
  // Rechtes Auge
  circle(180, 120, 20);
  
  // Mund (Linie)
  stroke(0);
  strokeWeight(3);
  line(120, 170, 180, 170);
}
````

````p5js
%show
````

### Demonstration 2: Ein animiertes Rechteck

````p5js
// Variable fÃ¼r Position
let x = 0;

function setup() {
  createCanvas(400, 200);
}

function draw() {
  background(220);
  
  // Rechteck, das sich nach rechts bewegt
  fill(100, 150, 255);
  rect(x, 75, 50, 50);
  
  // Position erhÃ¶hen
  x = x + 2;
  
  // Falls das Rechteck das Ende erreicht, zurÃ¼cksetzen
  if (x > 400) {
    x = 0;
  }
}
````

````p5js
%show
````

**Was passiert?**
- `let x = 0` deklariert eine Variable fÃ¼r die x-Position
- In `draw()` wird `x` jedes Frame erhÃ¶ht
- Das Rechteck bewegt sich nach rechts
- Wenn `x > 400` wird, springt das Rechteck zurÃ¼ck zu 0
